```python
from PIL import Image
import os
from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = './cachorro.png'

if os.path.isfile(path):
    img = Image.open(path)
    print("Imagem ok")
else:
    print("Arquivo não encontrado")
```

```python
def converter_grayscale(img_colorida):
    largura, altura = img_colorida.size
    img_tons_de_cinza = Image.new("L", (largura, altura))

    for y in range(altura):
        for x in range(largura):
            pixel = img_colorida.getpixel((x, y))
            # Média ponderada dos canais de cor (R, G, B)
            valor_tons_de_cinza = int(0.299 * pixel[0] + 0.587 * pixel[1] + 0.114 * pixel[2])
            img_tons_de_cinza.putpixel((x, y), valor_tons_de_cinza)


    return img_tons_de_cinza
```

```python
# Converter para tons de cinza
img_tons_de_cinza = converter_grayscale(img)

# Exibir a imagem em tons de cinza
display(img_tons_de_cinza)
```

```python
def converter_binary(img):
  largura, altura = img.size
  img_binaria = Image.new("1", (largura, altura))

  for y in range(altura):
    for x in range(largura):
      pixel = img.getpixel((x, y))

      # Se estiver em modo "L", apenas um valor de pixel
      valor_pixel = pixel if isinstance(pixel, int) else int((pixel[0] + pixel[1] + pixel[2]) / 3)

      if valor_pixel > 127:
        img_binaria.putpixel((x, y), 1)
      else:
        img_binaria.putpixel((x, y), 0)

  return img_binaria
```

```python
# Converter para imagem binaria
img_binaria = converter_binary(img_tons_de_cinza)

# Exibir a imagem binaria
display(img_binaria)
```
